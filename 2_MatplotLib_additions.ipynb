```python
# Adapted from https://www.python-course.eu/matplotlib_contour_plot.php
from matplotlib import pyplot as plt
from mpl_toolkits import mplot3d
import numpy as np
```

```python
np.linspace(1, 10, 4)
```

```python
# Creating a grid with np.meshgrid()
xs = np.linspace(0, 2, 2).astype(int)
ys = np.linspace(0, 2, 3).astype(int)
print('xs =', xs)
print('ys =', ys)
X, Y = np.meshgrid(xs, ys)
print(X.shape, Y.shape)
print('X =', X)
print('Y =', Y)
plt.figure(figsize=(2,2))
plt.scatter(X, Y, color="green")
plt.title('The grid created by meshgrid()')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
xs = np.linspace(-3.0, 3.0, 7)
ys = np.linspace(-3.0, 3.0, 7)
X, Y = np.meshgrid(xs, ys)
```

```python
Z = X**2 + Y**2
plt.figure(figsize=(7,5))
ax = plt.subplot(projection='3d')
ax.plot_surface(X, Y, Z)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
plt.show()
```

```python
fig = plt.figure(figsize=(6,5))

#ax = plt.subplot()
levels = [0, 1, 2, 4, 8, 13, 18]
cp = plt.contour(X, Y, Z, levels, colors='black', linestyles='dashed')
plt.clabel(cp, inline=True, fontsize=10)
plt.title('Contour Plot')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
